## 1.とりあえず、テキストを抽出する方法から。

```python
import bs4 as bs
import urllib.request

sauce = urllib.request.urlopen("https://pythonprogramming.net/parsememcparseface/").read()

soup = bs.BeautifulSoup(sauce, "lxml")


print(soup.find_all("p"))  # Pタグが付いているものをすべて表示(タグも表示)
print(soup.p.text)             # pタグが付いているものをすべて表示（テキストだけ）


for pg in soup.find_all("p"):   # for loopを使って抽出
    print(pg.text)

print(soup.get_text())  #すべてのテキストを表示
```

## 2. タグ別にテキストを抽出する 

```python
import bs4 as bs
import urllib.request

sauce = urllib.request.urlopen("https://pythonprogramming.net/parsememcparseface/").read()

soup = bs.BeautifulSoup(sauce, "lxml")

nav = soup.nav                        #navigationの情報をとる
for url in nav.find_all("a"):
    print(url.get("href"))
    
body = soup.body                 #<body>だけのの情報をとる
for pg in body.find_all("p"):
    print(pg.text)


for pg in soup.find_all("div", class_="body"):    #<div>の情報をとる: クラスも指定できる
    print(pg.text)
```

## 3. テーブルからデータを抽出

#### (a) bs4を使った方法

```python
import bs4 as bs
import urllib.request

sauce = urllib.request.urlopen("https://pythonprogramming.net/parsememcparseface/").read()
soup = bs.BeautifulSoup(sauce, "lxml")

table = soup.find("table")

table_rows = table.find_all("tr")

for tr in table_rows:
    td = tr.find_all("td")
    row =[i.text for i in td]
    print(row)
```

#### (b) pandasを使った方法（こっちの方がはるかに楽）

## 4. Sitemap（すべてのリンクがまとめられてるページ）からアドレスを抽出

```python
import requests
import bs4 as bs
import re

url = "https://www.amazon.com/gp/product/B01MCR69IN/ref=s9u_qpp_gw_i2?ie=UTF8&fpl=fresh&pd_rd_i=B01MCR69IN&pd_rd_r=ZMZYENW7G8CSWW01Y0M1&pd_rd_w=2wSW2&pd_rd_wg=UguQf&pf_rd_m=ATVPDKIKX0DER&pf_rd_s=&pf_rd_r=E0T94GQGG1TCEVSPRTDN&pf_rd_t=36701&pf_rd_p=2a4fafb6-9fdc-425a-aee8-c82daa7b18ed&pf_rd_i=desktop"
content = requests.get(url).content
soup = bs.BeautifulSoup(content,"html.parser")

 
#ASIN
bucket = soup.find("td", class_="bucket")
for i in bucket.find_all("li"):
    text = i.text
    if "ASIN" in text:
        print(text)



```
